# Projections of Images using Radon Transform

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.data import shepp_logan_phantom
from skimage.transform import radon, rescale
import pandas as pd
from skimage.transform import iradon
```

```python
rabbit = imread('rabbit.jpg')[:,:,0]
cow = imread('cow.jpg')[:,:,0]
pig = imread('pig.jpg')
dog = imread('dog.jpg')[:,:,0]
cock = imread('cock.jpg')[:,:,0]
coe = imread('coe.jpg')[:,:,0]
```

## Normal Images

```python
f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3,figsize = (12, 8))

ax1.set_title('rabbit')
ax1.imshow(rabbit, cmap = "Greys_r")

ax2.set_title('cow')
ax2.imshow(cow, cmap = "Greys_r")

ax3.set_title('pig')
ax3.imshow(pig, cmap = "Greys_r")

ax4.set_title('dog')
ax4.imshow(dog, cmap = "Greys_r")

ax5.set_title('cock')
ax5.imshow(cock, cmap = "Greys_r")

ax6.set_title('coe')
ax6.imshow(coe, cmap = "Greys_r")
```

## Sinograms of images at Zero degrees

```python
theta = [0.]
f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3,figsize = (12, 8))

ax1.set_title('rabbit sinogram')
#ax1.imshow(radon(rabbit, theta, circle = False ), cmap = "Greys_r")
ax1.plot(radon(rabbit, theta, circle = False ))

ax2.set_title('cow sinogram')
#ax2.imshow(radon(cow, theta, circle = False ), cmap = "Greys_r")
ax2.plot(radon(cow, theta, circle = False ))

ax3.set_title('pig sinogram')
#ax3.imshow(radon(pig, theta, circle = False ), cmap = "Greys_r")
ax3.plot(radon(pig, theta, circle = False ))

ax4.set_title('dog sinogram')
#ax4.imshow(radon(dog, theta, circle = False ), cmap = "Greys_r")
ax4.plot(radon(dog, theta, circle = False ))

ax5.set_title('cock sinogram')
#ax5.imshow(radon(cock, theta, circle = False ), cmap = "Greys_r")
ax5.plot(radon(cock, theta, circle = False ))

ax6.set_title('coe sinogram')
#ax6.imshow(radon(coe, theta, circle = False ), cmap = "Greys_r")
ax6.plot(radon(coe, theta, circle = False ))
```
