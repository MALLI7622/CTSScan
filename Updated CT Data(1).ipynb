```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import radon, iradon
import random
import tarfile
import pydicom
import dicom
import pandas as pd
import rarfile
import patoolib
import os
import cv2
```

```python
data = patoolib.extract_archive('CTBraindata.zip')
```

```python
data_dir = 'CTBraindata/202-brain 1X1-84670/'
```

```python
data = os.listdir(data_dir)
```

```python
theta = np.linspace(0., 180., 100)
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    i = random.randint(0, 40)
    data_dcm = pydicom.dcmread(data_dir + '/' + data[i])
    plt.imshow(data_dcm.pixel_array, cmap = "Greys_r")
plt.show()
```

```python
i = random.randint(0, 100)
sample = pydicom.dcmread(data_dir + '/' + data[i])
theta = np.linspace(0., 180., 180)

sinogram = radon(sample.pixel_array, theta = theta, circle = False)

inverted_sinogram = iradon(sinogram, theta = theta, circle = False)

f , (ax1, ax2, ax3) = plt.subplots(1,3 , figsize = (12, 6))

ax1.set_title(' Original Image')
ax1.imshow(sample.pixel_array, cmap = "Greys_r")

ax2.set_title(' Sinogram')
ax2.imshow(sinogram, cmap = "Greys_r")

ax3.set_title('Inverted Sinogram')
ax3.imshow(inverted_sinogram, cmap = "Greys_r")
```
